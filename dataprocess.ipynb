```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import os
```

```python
file_path = "data//org//vehicle.dat"
file_name='vehicle'
```

```python
# 使用read_csv函数加载.dat文件
df = pd.read_csv(file_path, sep=',', header=None) #sep=','
df.iloc[0,:]
```

```python
## usd for .data file
import pandas as pd

# Load the data using a regular expression for any whitespace
df = pd.read_csv(file_path, sep='\s+', header=None)

# Display the first row
df.iloc[0,:]
```

```python
df.head(20)
```

```python
# 对所有含字母的列进行 factorize 编码
for col in df.columns:
    if df[col].dtype == 'object':  # 检查列是否为离散字母
        df[col], _ = pd.factorize(df[col])

print(df)
```

```python
scaler = StandardScaler()

# 对每列数据进行标准化处理
df.iloc[:,:-1] = scaler.fit_transform(df.iloc[:,:-1])
```

```python
df
```

```python
last_column = df[df.columns[-1]]
value_counts = last_column.value_counts()
print(value_counts)
```

```python
train,val = train_test_split(df,test_size=0.2,random_state=4)
```

```python

process_file_path = 'data//now//'+file_name
if not os.path.exists(process_file_path):
    os.makedirs(process_file_path)
```

```python
train_file = os.path.join(process_file_path,'train.csv')
val_file = os.path.join(process_file_path,'val.csv')
```

```python
train.to_csv(train_file)
val.to_csv(val_file)
```
